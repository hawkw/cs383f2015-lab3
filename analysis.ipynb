100-agent runs
--------------

This includes the run with heterogenous agents. 

```python
runs100 = [ pandas.read_csv('results/elfarol-100agents-regression.csv'),
            pandas.read_csv('results/elfarol-100agents-random.csv'),
            pandas.read_csv('results/elfarol-100agents-average.csv'),
            pandas.read_csv('results/elfarol-100agents-markov.csv'),
            pandas.read_csv('results/elfarol-100agents-heterogenous.csv')]
df100 = pandas.concat(runs100, keys=['regression','random','averaging','markov','heterogenous'])
```

```python
df100 = df100.reset_index(0)
df100.rename(columns={"level_0":"Run Type"}, inplace=True)
df100.describe()
```

```python
sns.lmplot(x='tick', y='Attending Count', data=df100, hue='Run Type')
```

```python
g = sns.FacetGrid(df100, col="Run Type", col_wrap=2, margin_titles=True, size=7, hue="Run Type")
g.map(plt.scatter, "tick", "Attending Count")
```

For the initial set of 100-agent runs, all of the prediction strategies are relatively 'noisy', and determining patterns in agent behaviour is fairly difficult. However, we can see here the ineffectiveness of the averaging strategy, which simply alternates between the maximum and minimum attendance levels. This behaviour maximizes neither the utility of individual agents nor the social welfare of the entire system

1,000-agent runs
----------------

The runs with 1,000 and 10,000 agents are more interesting to look at, as they show more variance between the regression-based and Markov prediction agents.

```python
runs1000 = [ pandas.read_csv('results/elfarol-1000agents-regression.csv'),
            pandas.read_csv('results/elfarol-1000agents-random.csv'),
            pandas.read_csv('results/elfarol-1000agents-average.csv'),
            pandas.read_csv('results/elfarol-1000agents-markov.csv')]
df1000 = pandas.concat(runs1000, keys=['regression','random','averaging','markov'])
```

```python
df1000 = df1000.reset_index(0)
df1000.rename(columns={"level_0":"Run Type"}, inplace=True)
df1000.describe()
```

```python
g = sns.FacetGrid(df1000, col="Run Type", col_wrap=2, margin_titles=True, size=7, hue="Run Type")
g.map(plt.scatter, "tick", "Attending Count")
```

When the number of agents in the model is increased, patterns in the attendance levels begin to emerge. A pattern in which attendance alternates between five levels of attendance emerges in both the Markov chain-based prediction strategy and the regression-based predictive strategy, although this pattern emerges much earlier in the simulation for the regression-based strategy.

10,000-agent runs
-----------------

```python
runs10000 = [ pandas.read_csv('results/elfarol-10000agents-regression.csv'),
            pandas.read_csv('results/elfarol-10000agents-random.csv'),
            pandas.read_csv('results/elfarol-10000agents-average.csv'),
            pandas.read_csv('results/elfarol-10000agents-markov.csv')]
df10000 = pandas.concat(runs10000, keys=['regression','random','averaging','markov'])
```

```python
df10000 = df10000.reset_index(0)
df10000.rename(columns={"level_0":"Run Type"}, inplace=True)
df10000.describe()
```

```python
g = sns.FacetGrid(df10000, col="Run Type", col_wrap=2, margin_titles=True, size=7, hue="Run Type")
g.map(plt.scatter, "tick", "Attending Count")
```

Some interesting patterns are observable when the number of agents in the model is 10,000. 

For example, the attendance of agents using the random prediction strategy decreases rapidly over time. Agents using this strategy alternate between a high level of attendance, a low level of attendance, and zero attendance. This is due to the random value that predicts the next attendance level being from within the maximum remembered and minimum remembered attendance level.

At this number of agents in the model, the Markov chain-based prediction strategy begins to display some deterministic behaviour over the course of the run. While the attendance levels of agents using this strategy are initially widely scattered, after about 150 ticks of the simulation, the agents begin to alternate through four very similar levels of attendance.

Threshold=90 runs
-----------------

```python
runs90t = [ pandas.read_csv('results/elfarol-100agents-threshold90-regression.csv'),
            pandas.read_csv('results/elfarol-100agents-threshold90-random.csv'),
            pandas.read_csv('results/elfarol-100agents-threshold90-average.csv'),
            pandas.read_csv('results/elfarol-100agents-threshold90-markov.csv')]
df90t = pandas.concat(runs90t, keys=['regression','random','averaging','markov'])
```

```python
df90t = df90t.reset_index(0)
df90t.rename(columns={"level_0":"Run Type"}, inplace=True)
df90t.describe()
```

```python
g = sns.FacetGrid(df90t, col="Run Type", col_wrap=2, margin_titles=True, size=7, hue="Run Type")
g.map(plt.scatter, "tick", "Attending Count")
```

When the overcrowding threshold is closer to the number of agents in the model, the minimum and maximum attendance of agents following most strategies are brought to within a closer range. For example, the regression and Markov chain prediction strategies both appear to approach a stable level of attendance, although the regression-based strategy is able to make more accurate predictions in this scenario.
